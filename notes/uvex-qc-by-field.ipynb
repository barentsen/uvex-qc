# Status of the UVEX survey

```python
# Setup the notebook
%pylab inline
import matplotlib.pyplot as plt
from astropy.table import Table
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in the observing data
logs = Table.read('data/int-logs/uvex-unreduced-fields.fits')
casu = Table.read('data/casu-dqc/uvex-casu-dqc-by-field.fits')
```

## What has been observed so far?

```python
n_fields = 7635*2
n_fields_observed = unique(concatenate((casu['field'], logs['field']))).size
print('Unique fields observed: {0}/{1} ({2:.1f}%)'.format(n_fields_observed, n_fields, 100*n_fields_observed/float(n_fields)))
```

## What has been *reduced* so far?

```python
n_fields_reduced = unique(casu['field']).size
print('Unique fields reduced: {0}/{1} ({2:.1f}%)'.format(n_fields_reduced, n_fields, 100*n_fields_reduced/float(n_fields)))
```

## What is the quality of the reduced data?

### PSF FWHM

```python
medians = [median(casu['seeing_'+band][-isnan(casu['seeing_'+band])]) for band in ['u', 'g', 'r', 'hei']]
print('Median FWHM = {0:.2f}" (U), {1:.2f}" (g), {2:.2f}" (r), {3:.2f}" (HeI).'.format(*medians))

figure(figsize=(20,10))
subplot(221)
hist(casu['seeing_u'], range=[0.5,3], bins=25)
xlabel('U-band FWHM [arcsec]')
subplot(222)
hist(casu['seeing_g'], range=[0.5,3], bins=25)
xlabel('g-band FWHM [arcsec]')
subplot(223)
hist(casu['seeing_r'], range=[0.5,3], bins=25)
xlabel('r-band FWHM [arcsec]')
subplot(224)
hist(casu['seeing_hei'], range=[0.5,3], bins=25)
xlabel('HeI-band FWHM [arcsec]')
show()
```

```python
figure(figsize=(20,3))
scatter(casu['l'], casu['b'], c=casu['seeing_u'], edgecolor='none', alpha=0.75, s=10, vmin=0.75, vmax=2.25)
xlabel('Galactic longitude (l)')
ylabel('Galactic latitude (b)')
xlim([220, 20])
ylim([-6, +6])
colorbar(label='U-band PSF FWHM [arcsec]')
show()
```

### Airmass

```python
medians = [median(casu['airmass_'+band][-isnan(casu['airmass_'+band])]) for band in ['u', 'g', 'r', 'hei']]
print('Median airmass = {0:.2f}" (U), {1:.2f}" (g), {2:.2f}" (r), {3:.2f}" (HeI).'.format(*medians))

figure(figsize=(6,5))
hist(casu['airmass_u'], range=[1,2], bins=50)
xlabel('U-band airmass')
show()
```

### Ellipticity

```python
medians = [median(casu['ellipt_'+band][-isnan(casu['ellipt_'+band])]) for band in ['u', 'g', 'r', 'hei']]
print('Median ellipticity = {0:.2f} (U), {1:.2f} (g), {2:.2f} (r), {3:.2f} (HeI).'.format(*medians))

figure(figsize=(20,10))
subplot(221)
hist(casu['ellipt_u'], range=[0,0.3], bins=12)
xlabel('U-band ellipticity')
subplot(222)
hist(casu['ellipt_g'], range=[0,0.3], bins=12)
xlabel('g-band ellipticity')
subplot(223)
hist(casu['ellipt_r'], range=[0,0.3], bins=12)
xlabel('r-band ellipticity')
subplot(224)
hist(casu['ellipt_hei'], range=[0,0.3], bins=12)
xlabel('r-band ellipticity')
show()
```

### Sky level

```python
medians = [median(casu['sky_'+band][-isnan(casu['sky_'+band])]) for band in ['u', 'g', 'r', 'hei']]
print('Median sky level: {0:.0f} (U), {1:.0f} (g), {2:.0f} (r), {3:.0f} (HeI).'.format(*medians))

figure(figsize=(20,10))
subplot(221)
hist(casu['sky_u'], range=[0,250], bins=25)
xlabel('U-band sky level [counts]')
subplot(222)
hist(casu['sky_g'], range=[0,250], bins=25)
xlabel('g-band sky level [counts]')
subplot(223)
hist(casu['sky_r'], range=[0,250], bins=25)
xlabel('r-band sky level [counts]')
subplot(224)
hist(casu['sky_hei'], range=[0,250], bins=25)
xlabel('HeI-band sky level [counts]')
show()
```

```python
SEEING_LIMIT = 2.0
ELLIPT_LIMIT = 0.3
AIRMASS_LIMIT = 1.4
good = ((casu['seeing_u'] < SEEING_LIMIT)
        & (casu['seeing_g'] < SEEING_LIMIT)
        & (casu['seeing_r'] < SEEING_LIMIT)
        & (casu['seeing_hei'] < SEEING_LIMIT)
        & (casu['ellipt_u'] < ELLIPT_LIMIT)
        & (casu['ellipt_g'] < ELLIPT_LIMIT)
        & (casu['ellipt_r'] < ELLIPT_LIMIT)
        & (casu['ellipt_hei'] < ELLIPT_LIMIT)
        & (casu['airmass_u'] < AIRMASS_LIMIT)
        & (casu['airmass_g'] < AIRMASS_LIMIT)
        & (casu['airmass_r'] < AIRMASS_LIMIT)
        & (casu['airmass_hei'] < AIRMASS_LIMIT)
        )
print('{0} {1}'.format(unique(casu['field'][good]).size, unique(casu['field']).size))
```
