# Status of the UVEX survey

```python
# Setup the notebook
%pylab inline
import matplotlib.pyplot as plt
from astropy.table import Table
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in the observing data
logs = Table.read('data/int-logs/uvex-logs-by-run.fits')
casu = Table.read('data/casu-dqc/uvex-casu-dqc-by-run.fits')
# Create a mask for each filter
logs_u = np.char.strip(logs['filter']) == 'U'
logs_g = np.char.strip(logs['filter']) == 'g'
logs_r = np.char.strip(logs['filter']) == 'r'
casu_u = np.char.strip(casu['filter']) == 'U'
casu_g = np.char.strip(casu['filter']) == 'g'
casu_r = np.char.strip(casu['filter']) == 'r'
```

## What has been observed so far?

```python
n_fields = 7635*2
n_fields_observed = unique(logs['field']).size
print('Unique fields observed: {0}/{1} ({2:.1f}%)'.format(n_fields_observed, n_fields, 100*n_fields_observed/float(n_fields)))
```

```python
figure(figsize=(20,3))
scatter(logs['l'][logs_r], logs['b'][logs_r], facecolor='black', edgecolor='none', alpha=0.5, s=10)
xlim([220, 25])
ylim([-6, +6])
show()
```

## What has been *reduced* so far?

```python
n_fields_reduced = unique(casu['field']).size
print('Unique fields reduced: {0}/{1} ({2:.1f}%)'.format(n_fields_reduced, n_fields, 100*n_fields_reduced/float(n_fields)))
```

```python
figure(figsize=(20,3))
scatter(casu['l'][casu_r], casu['b'][casu_r], facecolor='black', edgecolor='none', alpha=0.5, s=10)
xlim([220, 20])
ylim([-6, +6])
show()
```

## What is the quality of the reduced data?

### PSF FWHM

```python
medians = [median(casu['seeing'][mask]) for mask in [casu_u, casu_g, casu_r]]
print('Median FWHM = {0:.2f}" (U), {1:.2f}" (g), {2:.2f}" (r).'.format(*medians))

figure(figsize=(20,5))
subplot(131)
hist(casu['seeing'][casu_u], range=[0.5,3], bins=25)
xlabel('U-band FWHM [arcsec]')
subplot(132)
hist(casu['seeing'][casu_g], range=[0.5,3], bins=25)
xlabel('g-band FWHM [arcsec]')
subplot(133)
hist(casu['seeing'][casu_r], range=[0.5,3], bins=25)
xlabel('r-band FWHM [arcsec]')
show()
```

```python
figure(figsize=(20,3))
scatter(casu['l'][casu_r], casu['b'][casu_r], c=casu['seeing'][casu_r], edgecolor='none', alpha=0.75, s=10, vmin=0.5, vmax=2.5)
xlim([220, 20])
ylim([-6, +6])
colorbar(label='r-band seeing [arcsec]')
show()
```

### Ellipticity

```python
medians = [median(casu['ellipt'][mask]) for mask in [casu_u, casu_g, casu_r]]
print('Median ellipticity = {0:.2f} (U), {1:.2f} (g), {2:.2f} (r).'.format(*medians))

figure(figsize=(20,5))
subplot(131)
hist(casu['ellipt'][casu_u], range=[0,0.3], bins=12)
xlabel('U-band ellipticity')
subplot(132)
hist(casu['ellipt'][casu_g], range=[0,0.3], bins=12)
xlabel('g-band ellipticity')
subplot(133)
hist(casu['ellipt'][casu_r], range=[0,0.3], bins=12)
xlabel('r-band ellipticity')
show()
```

### Sky level

```python
medians = [median(casu['sky'][mask]) for mask in [casu_u, casu_g, casu_r]]
print('Median sky = {0:.0f} (U), {1:.0f} (g), {2:.0f} (r).'.format(*medians))

figure(figsize=(20,5))
subplot(131)
hist(casu['sky'][casu_u], range=[0,250], bins=25)
xlabel('U-band sky level [counts]')
subplot(132)
hist(casu['sky'][casu_g], range=[0,250], bins=25)
xlabel('g-band sky level [counts]')
subplot(133)
hist(casu['sky'][casu_r], range=[0,250], bins=25)
xlabel('r-band sky level [counts]')
show()
```

### Airmass

```python
medians =[median(casu['airmass'][mask]) for mask in [casu_u, casu_g, casu_r]]
print('Median airmass = {0:.2f} (U), {1:.2f} (g), {2:.2f} (r).'.format(*medians))

figure(figsize=(6,5))
hist(casu['airmass'][casu_u], range=[1,2], bins=50)
xlabel('U-band airmass')
show()
```
